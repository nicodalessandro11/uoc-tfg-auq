```python
import requests
import pandas as pd

# Resource ID del CSV que quieres (reemplaza por el tuyo)
RESOURCE_ID = "21f7a4df-2e73-45f8-8c6d-0b3db8c21527"

# URL base
BASE_URL = "https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search"

# Descarga paginada
limit = 1000
offset = 0
all_records = []

while True:
    params = {
        "resource_id": RESOURCE_ID,
        "limit": limit,
        "offset": offset
    }
    response = requests.get(BASE_URL, params=params)
    data = response.json()

    # A帽ade los resultados a la lista
    records = data["result"]["records"]
    if not records:
        break
    all_records.extend(records)
    
    # Avanza la paginaci贸n
    offset += limit

# Carga a un DataFrame y guarda a CSV
df = pd.DataFrame(all_records)
df.to_csv("dataset_barcelona.csv", index=False)
print(f"Guardado {len(df)} registros en 'dataset_barcelona.csv'")
```
